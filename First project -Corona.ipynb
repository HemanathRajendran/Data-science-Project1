```python
import pandas as pd
import csv
from datetime import date, timedelta
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv('corona.csv',usecols = ['Country/Region','Confirmed','ObservationDate'])
confirmed=file[file['Country/Region']=='India'] ['Confirmed']
lists=confirmed.values.tolist()
update_values=[887,987,1024,1251,1397,1998,2543]
data=lists + update_values
print(data)
length=len(data)
print('length=',length)
total_cs=0
for i in range(length-1):
    cs=data[i+1]/data[i]
    total_cs+=cs
rate_of_growth=total_cs/length
print('Rate of growth=',rate_of_growth)
cases=[]
for j in range(13):
    y= 2543*(rate_of_growth)**(j+1)
    cases.append(y)
print(cases)
date=['4/3/2020','4/4/2020','4/5/2020','4/6/2020','4/7/2020','4/8/2020','4/9/2020','4/10/2020','4/11/2020','4/12/2020','4/13/2020','4/14/2020','4/15/2020']
fig = plt.figure()
axes = fig.add_axes([0.5,0.5,2,2])
plt.plot(date,cases,'r') 
plt.xlabel('DATE')
plt.ylabel('CASES')
plt.title('INDIA COVID-19')
plt.show()
```
